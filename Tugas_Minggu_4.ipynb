```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/foto.jpg' , cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
plt.title('Tugas 1(Grayscale)')
plt.show()

Binary =cv2.threshold(img,127,255,cv2.THRESH_BINARY)[1]
plt.imshow(Binary,cmap='gray')
plt.title('Tugas 1(Grayscale)')
plt.show()
```

```python
negative = 255 - img
plt.imshow(negative,cmap='gray')
plt.title('Tugas 2(Negative).')
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img_rgb = cv2.imread('/foto.jpg')

img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)

c = 255 / (np.log(1 + np.max(img_gray)))
log_image_gray = c * (np.log(img_gray + 1))
log_image_gray = np.array(log_image_gray, dtype = np.uint8)

plt.imshow(log_image_gray,cmap='gray')
plt.title('Tugas 3.')
plt.show()
```

```python
gamma = 0.5
c = 1
power_law = c * np.power(img / 255.0, gamma)
power_law = np.uint8(power_law * 255)

plt.imshow(img,cmap='gray')
plt.title('Tugas 4')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/foto.jpg',cv2.IMREAD_GRAYSCALE)

Kernel = np.array([[-0,-1,-0],[-1,5,-1],[0,-1,0]])

conv = cv2.filter2D(img,-1,Kernel)

plt.subplot(1,2,1), plt.imshow(img,cmap='gray'),plt.title('Tugas 5.')
plt.subplot(1,2,2), plt.imshow(conv,cmap='gray'),plt.title('Tugas 5.')
plt.show()
```
